```python
import sqlite3
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
pd.options.mode.chained_assignment = None
```

## Female Data Visualization

```python
data = pd.read_csv('female.csv', low_memory=False)
data
```

```python
# clean data
# new dataframe to minimize data loss due to cleaning
height_score = pd.DataFrame()
height_score ['height (cm)'] = data.loc[:, 'height (cm)']
height_score ['score'] = data.loc[:,'score']
height_score = height_score.dropna()
height_score
```

```python
# visualize height and score average
x = height_score.loc[:, 'height (cm)']
y = height_score.loc[:,'score']

plt.scatter(x,y)
plt.title('Female Height and Score average')
plt.show()
```

```python
# clean data
length_score = pd.DataFrame()
length_score ['climbing length'] = data.loc[:, 'climbing length']
length_score.drop(length_score[length_score['climbing length'] >= 2022].index, inplace = True)
length_score ['score'] = data.loc[:,'score']
length_score = length_score.dropna()

# visualize height and score average
x = length_score.loc[:, 'climbing length']
y = length_score.loc[:,'score']

plt.scatter(x,y)
plt.title('Female Climbing length (years) and Score average')
plt.show()
```

```python
# clean data
age_score = pd.DataFrame()
age_score ['age'] = data.loc[:, 'age']
age_score.drop(age_score[age_score['age'] >= 2022].index, inplace = True)
age_score ['score'] = data.loc[:,'score']
age_score = age_score.dropna()

# visualize height and score average
x = age_score.loc[:, 'age']
y = age_score.loc[:,'score']

plt.scatter(x,y)
plt.title('Female Age (years) and Score average')
plt.show()
```

```python
# clean data for tableau visualization
data.drop(data[data['age'] >= 2022].index, inplace = True)
data.drop(data[data['climbing length'] >= 2022].index, inplace = True)
data = data.dropna()
data.to_csv('female_boulderers.csv')
data
```

```python
import seaborn as sns
import numpy as np
from sklearn import datasets

db = pd.DataFrame.from_records(data, columns=['score','height (cm)','climbing length', 'age'])

#pd.plotting.scatter_matrix(db) 
sns.pairplot(db) 
plt.show()
```
